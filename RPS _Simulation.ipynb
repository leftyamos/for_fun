```python
import random

import matplotlib.pyplot as plt
```

```python
# rock, paper, scissors choices
choices = ['rock', 'paper', 'scissors']
```

### CODE: Monte Carlo Simulation for "Rock, Paper, Scissors" 

```python
def rps(n, player_choice):
    
    # initiate scores
    player_score = 0
    pc_score = 0
    
    # loop for rock, paper, scissors game
    for i in range(n):
        # computer choice
        pc = choices[random.randint(0,2)]
    
        # player choice
        player = player_choice
    
    
        # outcome tie
        if player == pc:
            None

        # outcome rock
        elif player == 'rock':
            if pc == 'paper':
                pc_score += 1
            elif pc == 'scissors':
                player_score += 1

        # outcome paper
        elif player == 'paper':
            if pc == 'rock':
                player_score += 1
            elif pc == 'scissors':
                pc_score += 1

        # outcome scissors
        elif player == 'scissors':
            if pc == 'rock':
                pc_score += 1
            elif pc == 'paper':
                player_score += 1
        
        
        # append probability
        player_prob.append(player_score/(i+1))
        pc_prob.append(pc_score/(i+1))
        
        
    # plot 
    plt.axhline(y = 1/3, color = 'r')     ## reference line 
    plt.xlabel('Iteration')
    plt.ylabel('Win probability')
    plt.plot(player_prob, label = 'Player')
    plt.plot(pc_prob, label = 'Computer')
    plt.legend()
    
    
    # results in percentage
    print(f'Player win probability = {player_prob[-1]}%')
    print(f'Computer win probability = {pc_prob[-1]}%')
```

### What is the win probability if only rock is chosen for 5000 times?

```python
# initiate empty list to store the win probability for player & pc
player_prob = []
pc_prob = []

rps(5000, 'rock')
```

### What is the win probability if both player and computer choice are random?

```python
player_prob = []
pc_prob = []

rps(5000, choices[random.randint(0,2)])
```
